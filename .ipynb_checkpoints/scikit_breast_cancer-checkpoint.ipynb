```python
# ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã®ãƒ­ãƒ¼ãƒ‰
original_df = pd.read_csv('data.csv')
print('original_df shape: {}'.format(original_df.shape))
# original_df shape: (569, 33)

# ã‚´ãƒŸãƒ‡ãƒ¼ã‚¿ã®å‰Šé™¤
original_df.drop('Unnamed: 32', axis=1, inplace=True)
original_df
```

```python
original_df.diagnosis.value_counts()
# B    357
# M    212
# Name: diagnosis, dtype: int64

# ç›®çš„å¤‰æ•°ã®æŠ½å‡º
y = original_df.diagnosis.apply(lambda d: 1 if d == 'M' else 0)

# èª¬æ˜Žå¤‰æ•°ã®æŠ½å‡º
X = original_df.loc[:, 'radius_mean':]
```

```python
# å­¦ç¿’ç”¨ã¨ãƒ†ã‚¹ãƒˆç”¨ã§ãƒ‡ãƒ¼ã‚¿ã‚’åˆ†é›¢
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# æ¨™æº–åŒ–
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ãƒ­ã‚¸ã‚¹ãƒ†ã‚£ãƒƒã‚¯å›žå¸°ã§å­¦ç¿’
logistic = LogisticRegressionCV(cv=10, random_state=0)
logistic.fit(X_train_scaled, y_train)

# æ¤œè¨¼
print('Train score: {:.3f}'.format(logistic.score(X_train_scaled, y_train)))
print('Test score: {:.3f}'.format(logistic.score(X_test_scaled, y_test)))
print('Confustion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=logistic.predict(X_test_scaled))))
# Train score: 0.988
# Test score: 0.972
# Confustion matrix:
# [[89  1]
#  [ 3 50]]
```

```python
plt.figure(figsize=(20, 20))
seaborn.heatmap(pd.DataFrame(X_train_scaled).corr(), annot=True)
```

```python
pca = PCA(n_components=30)
pca.fit(X_train_scaled)
plt.bar([n for n in range(1, len(pca.explained_variance_ratio_)+1)], pca.explained_variance_ratio_)
```

```python
# PCA
# æ¬¡å…ƒæ•°2ã¾ã§åœ§ç¸®
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
print('X_train_pca shape: {}'.format(X_train_pca.shape))
# X_train_pca shape: (426, 2)

# å¯„ä¸ŽçŽ‡
print('explained variance ratio: {}'.format(pca.explained_variance_ratio_))
# explained variance ratio: [ 0.43315126  0.19586506]

# æ•£å¸ƒå›³ã«ãƒ—ãƒ­ãƒƒãƒˆ
temp = pd.DataFrame(X_train_pca)
temp['Outcome'] = y_train.values
b = temp[temp['Outcome'] == 0]
m = temp[temp['Outcome'] == 1]
plt.scatter(x=b[0], y=b[1], marker='o') # è‰¯æ€§ã¯â—‹ã§ãƒžãƒ¼ã‚¯
plt.scatter(x=m[0], y=m[1], marker='^') # æ‚ªæ€§ã¯â–³ã§ãƒžãƒ¼ã‚¯
plt.xlabel('PC 1') # ç¬¬1ä¸»æˆåˆ†ã‚’xè»¸
plt.ylabel('PC 2') # ç¬¬2ä¸»æˆåˆ†ã‚’yè»¸
```

```python
# ãƒ‘ã‚¤ãƒ—ãƒ©ã‚¤ãƒ³ã®ä½œæˆ
pca_pipeline = Pipeline([
    ('scale', StandardScaler()),
    ('decomposition', PCA(n_components=2)),
    ('model', LogisticRegressionCV(cv=10, random_state=0))
])

# æ¨™æº–åŒ–ãƒ»æ¬¡å…ƒåœ§ç¸®ãƒ»å­¦ç¿’
pca_pipeline.fit(X_train, y_train)

# æ¤œè¨¼
print('Train score: {:.3f}'.format(pca_pipeline.score(X_train, y_train)))
print('Test score: {:.3f}'.format(pca_pipeline.score(X_test, y_test)))
print('Confustion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=pca_pipeline.predict(X_test))))
# Train score: 0.965
# Test score: 0.937
# Confustion matrix:
# [[84  6]
#  [ 3 50]]
```
